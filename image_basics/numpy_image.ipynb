```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline 
```

```python
pic = Image.open('../Resources/Computer-Vision-with-Python/DATA/00-puppy.jpg') # path relative to notebook
type(pic)
```

```python
# transform picture into array
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
# red channel values 0-255
plt.imshow(pic_red[:,:,0], cmap='gray') # show this in grayscale
```

```python
pic_green = pic_arr.copy()
# green channel values 0-255
plt.imshow(pic_green[:,:,1], cmap='gray') # show this in grayscale
```

```python
pic_blue = pic_arr.copy()
# blue channel values 0-255
plt.imshow(pic_blue[:,:,2], cmap='gray') # show this in grayscale
```

As we can see, the different channels are slightly different in grayscale. The brighter it is, the higher the value of pixels in that channel. In our picture, the blue channel is the darkest which means our image has less blue than green or red.

```python
# zero out the green channel
pic_red[:,:,1] = 0
plt.imshow(pic_red)
```

Since we removed the green channel, the only colors acting on the image are red and blue which makes purple

```python
# zero out the blue channel
pic_red[:,:,2] = 0
plt.imshow(pic_red)
```
