```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_image = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image, pt1=(384,10), pt2=(500, 150), color=(0,255,0), thickness=10)
plt.imshow(blank_image)
```

Draws onto the original image permanently

```python
cv2.rectangle(blank_image, pt1=(200,200), pt2=(300, 300), color=(0,0,255), thickness=10)
plt.imshow(blank_image)
```

```python
cv2.circle(blank_image, center=(100,100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(blank_image)
```

```python
cv2.circle(blank_image, center=(400,400), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_image)
```

```python
cv2.line(blank_image, (0,0), (512,512), (102,255,255), 5)
plt.imshow(blank_image)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_image, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255,255,255), 
            thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_image)
```
