```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../Resources/Computer-Vision-with-Python/DATA/00-puppy.jpg")
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img)
```
