```python
# Additional param 0 makes it gray scale.
img = cv2.imread("../Resources/Computer-Vision-with-Python/DATA/rainbow.jpg", 0)
plt.imshow(img, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(img, 255//2, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh1, cmap='gray')
```

```python
img = cv2.imread("../Resources/Computer-Vision-with-Python/DATA/crossword.jpg", 0)
plt.imshow(img, cmap='gray')
```

Displaying images as figures in Jupyter



```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

https://en.wikipedia.org/wiki/Thresholding_(image_processing)

```python
# Adaptive threshold
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
# blending the two results from th1 and th2 
blended = cv2.addWeighted(th1, 0.6, th2, 0.4, 0)
show_pic(blended)
```
