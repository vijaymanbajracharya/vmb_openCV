```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
gorilla = cv2.imread(f"../Resources/Computer-Vision-with-Python/DATA/gorilla.jpg", 0)
```

```python
display(gorilla, cmap='gray')
```
